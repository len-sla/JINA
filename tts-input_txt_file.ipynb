````python
def tts_de(text, filename, directory):
  audio = model.apply_tts(texts=[text],
                        sample_rate=sample_rate)
  direct_and_filename= directory+'/'+filename
  with open(direct_and_filename, 'wb') as f:
    f.write(Audio(audio[0], rate=sample_rate).data)
    print(direct_and_filename, '  was saved\n')
````

````python
def folder_check(path):
    try_num = 1
    oripath = path[:-1] if path.endswith('/') else path
    if(not os.path.exists(path)):
      os.mkdir(path)
      print(path, '   created')
    else:    
      while os.path.exists(path):
          print("Delete content of  existing folder " + path + "?(Y/N)")
          decision = input()
          if decision == "Y":
              shutil.rmtree(path, ignore_errors=True)
              os.mkdir(path)

              break
          else:
            path = oripath + "_%d/"%try_num
            try_num += 1
            if(not os.path.exists(path)):
              print("Create new folder " + path + "?(Y/N)")
              decision = input()
              if decision == "Y":
                os.mkdir(path)
                print(path, 'was created')
                break
      return path
````

````python
# generating file name with timestamp info for later comparison
def gen_file_with_timestamp(name='name', ext='mp4'):


    timestamp = dt.datetime.now()
    
    ts1 = timestamp.strftime("%A%d%B%Y%m")
    ts2 = timestamp.strftime("%H")
    ts3 = timestamp.strftime("%M")
    ts4 = timestamp.strftime("%S")
    file_name= name+ '_'+ ts1 +"-"+ ts2+"-"+ ts3+"-"+ ts4+ext
    return file_name
````
